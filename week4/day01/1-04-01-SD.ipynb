**5.** Нами закуплен новый станок для работы по металлу. Известно, что время работы этого станка до первого отказа подчиняется показательному(экспоненциальному) распределению со средним значением 1 год. Отказ этого станка для нас это огромные убытки. 

* Какова вероятность, что до первого отказа пройдет не менее 2 лет?

> ответ тут

$P(X≥2) = 0.135$

* Визуализируй это распределение

```python
import matplotlib.pyplot as plt
from scipy.stats import expon
import numpy as np

dist = expon(scale=1)
t = np.linspace(0, 5)
pdf = dist.pdf(t)


plt.figure(figsize=(7, 4))
plt.plot(t, pdf, "b", lw=2, label="PDF (λ=1)")

plt.axvline(2, color="red", linestyle="--", label="t = 2 года")

plt.title("Экспоненциальное распределение (PDF)")
plt.xlabel("t (годы)")
plt.ylabel("P")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

**7.** Рост мальчиков возрастной группы 15 лет - нормально распределённая случайная величина $X$ с параметрами $\mu=161$ см и $\sigma=4$ см.
* Какую долю костюмов для мальчиков, имеющих рост от 152 до 158 см, нужно предусмотреть в объёме производства для данной возрастной группы?

```python
import scipy.stats as st

st.norm.cdf(158, 161, 4) - st.norm.cdf(152, 161, 4)
```

> ответ тут

$P ≈ 0.214$

* Сформулируй правило трёх сигм для случайной величины $X\sim\mathcal{N}$.

> ответ тут

При нормальном распределении будет 3 стандартных отклонения: 68%, 95% и 99.7%

Т.е. почти все значения назодятся не дальше, чем на 3 стандартных отклонения

* Визуализируй следующие элементы: 
    * кривую плотности вероятности для указанных параметров и распределения
    * залей выдели цветом нужный фрагмет диаграммы (примерно это должно выглядеть [так](https://i.stack.imgur.com/c8F3F.png), только с твоими параметрами)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 161
sigma = 4
x = np.linspace(145, 175, 600)
pdf = norm.pdf(x, loc=mu, scale=sigma)
a, b = 152, 158
ya = norm.pdf(a, mu, sigma)
yb = norm.pdf(b, mu, sigma)

plt.plot(x, pdf, lw=2, label="PDF N(161, 4²)")
x_fill = np.linspace(a, b, 300)
plt.fill_between(x_fill, norm.pdf(x_fill, mu, sigma), alpha=0.35, label="152 ≤ X ≤ 158")
plt.vlines([a, b], ymin=0, ymax=[ya, yb], colors="red", linestyles="--", linewidth=2)

plt.title("Нормальное распределение (μ=161, σ=4)")
plt.xlabel("Рост, см")
plt.ylabel("Плотность вероятности")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.show()
```

❓От чего зависит форма кривой нормального распределения? 

> ответ тут

От среднего значения и стандартного отклонения. Среднее значение двигает центр "колокола", а стандартное отклонение изменяет его ширину или "растягивает" его.

**8**. Среднее время обслуживания одного клиента составляет 10 минут, а стандартное отклонение равно 2 минутам. Выстроилось очередь из 20 клиентов. Какова вероятность, что среднее время обсулуживание этих клиентов будет меньше 11 минут

`Подсказка: Центральная предельная теорема`

```python
mu, sigma, n = 10, 2, 20
se = sigma / np.sqrt(n)
p_mean = st.norm.cdf(11, loc=mu, scale=se)
print(p_mean)
```

Ответ: $0.987$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st


x = np.linspace(9, 12, 20)
pdf = st.norm.pdf(x, loc=mu, scale=se)

plt.plot(x, pdf, "b", lw=2, label=r"$N(10,\ (2/\sqrt{20})^2)$")

x_fill = np.linspace(min(x), 11, 300)
plt.fill_between(
    x_fill,
    st.norm.pdf(x_fill, mu, se),
    alpha=0.5,
    color="skyblue",
    label=rf"$P(\bar X < 11) \approx {p_mean:.3f}$",
)

plt.axvline(11, color="red", linestyle="--", lw=2, label="11 минут")

plt.title("Распределение среднего времени обслуживания (20 клиентов)")
plt.xlabel("Среднее время, мин")
plt.ylabel("Плотность вероятности")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
